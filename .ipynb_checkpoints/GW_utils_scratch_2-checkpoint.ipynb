```python
import numpy as np
import pandas as pd
from river_dl.gw_utils import amp_phi
```

```python
import os
```

```python
GW_data = os.path.join("DRB_output","GW.npz")
trn_data = os.path.join("DRB_output","trn_preds.feather")
tst_data = os.path.join("DRB_output","tst_preds.feather")
```

```python
gw_obs = np.load(GW_data)
```

```python
trn_preds = pd.read_feather(trn_data)
tst_preds = pd.read_feather(tst_data)
```

```python
def calc_amp_phi(thisData):
    segList = np.unique(thisData['seg_id_nat'])
    water_amp_preds = []
    water_phi_preds = []
    for thisSeg in segList:
        amp, phi = amp_phi(thisData.loc[thisData.seg_id_nat==thisSeg,"date"].values,thisData.loc[thisData.seg_id_nat==thisSeg,"seg_tave_water"])
        water_amp_preds.append(amp)
        water_phi_preds.append(phi)
    return pd.DataFrame({'seg_id_nat':segList,'water_amp_preds':water_amp_preds,'water_phi_preds':water_phi_preds})
```

```python
gw_trn = calc_amp_phi(trn_preds)
gw_tst = calc_amp_phi(tst_preds)
```

```python
def merge_pred_obs(obs_col,pred):
    obsDF = pd.DataFrame(gw_obs[obs_col],columns=gw_obs['GW_cols'])
    obsDF = obsDF.merge(pred)
    obsDF['Ar_preds']=obsDF['water_amp_preds']/obsDF['air_amp']
    obsDF['delPhi_preds'] = (obsDF['water_phi_preds']-obsDF['air_phi'])*365/(2*math.pi)
    return obsDF
```

```python
gw_trn_DF = merge_pred_obs('GW_trn',gw_trn)
gw_tst_DF = merge_pred_obs('GW_tst',gw_tst)
```
